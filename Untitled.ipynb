```python
import numpy as np
import matplotlib.pyplot as plt
from myplotting import plot_quantiles, HandlerQuantile, QuantileContainer
```

```python
labels = np.random.choice(5, size=100)
values = np.random.normal(size=100)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5), dpi=300)
plot_quantiles(labels, values, pct_inner=10, alpha=1, ax=axes[1], label='goodbye')
plot_quantiles(labels - 1, values, alpha=1, pct_outer=5, ax=axes[0], color='C1', orientation='vertical', label='hello')
axes[0].legend(handler_map={QuantileContainer: HandlerQuantile()})
plt.show()
```
