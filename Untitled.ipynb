```python
import numpy as np
import matplotlib.pyplot as plt
from bizplots import plot_quantiles, plot_spaghetti, plot_ribbons, handler_map
```

```python
labels = np.random.choice(5, size=1000)
values = np.random.normal(size=1000)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
plot_quantiles(labels, values, pct_inner=25, alpha=1, ax=axes[1], label='goodbye')
plot_quantiles(labels - 1, values, alpha=1, pct_outer=75, ax=axes[0], color='C1', orientation='horizontal', label='hello')
axes[0].axhline(0.5, linestyle=":", label='hline')
axes[0].legend(handler_map=handler_map)
axes[1].legend(handler_map=handler_map)
plt.show()
```

```python
timesteps = np.linspace(0, 10, num=1000)
a_samples = np.random.normal(0, 0.2, size=400)
b_samples = np.random.normal(loc=1, scale=0.1, size=400)

line_samples = a_samples[:, None] + b_samples[:, None] * timesteps
```

```python
fig, ax = plt.subplots()
plot_spaghetti(timesteps, line_samples, alpha=0.4, label='spag', ax=ax)
ax.legend(handler_map=handler_map)
plt.show()
```

```python
fig, ax = plt.subplots()
plot_ribbons(timesteps, line_samples, num_ribbons=3, label='spag', ax=ax)
ax.legend(handler_map=handler_map)
plt.show()
```

```python
import matplotlib.colors as mcolors
from matplotlib.container import Container

class RibbonsContainer(Container):
    def __init__(
        self,
        collections,
        **kwargs,
    ):
        self.collections = collections
        super().__init__(collections, **kwargs)


def plot_ribbons(
    x,
    y,
    num_ribbons=10,
    percentile_min=2.5,
    percentile_max=97.5,
    ax=None,
    **kwargs,
):
    """
    Make a ribbon plot that shows the different quantiles of the
    distribution of y against x.

    Parameters
    ----------
        x: 1d array
            The values for the x axis
        y: 2d array
        num_ribbons: int (default 10)
            How many quantiles to show
        percentile_min: float, between 0 and 50
            The lowest percentile to be shown
        percentile_max: float between 50 and 100
            The highest percentile to show.
        ax: matplotlib.Axes
            Where to plot the figure.
        kwargs: dict
            Extra arguments passed to `plt.fill_between`.
            Controls the aspect of the ribbons.

    Returns
    -------
        matplotlib.Axes
    """
    if ax is None:
        ax = plt.gca()
    
    lower = np.linspace(percentile_min, 50, num=num_ribbons, endpoint=False)
    upper = np.linspace(50, percentile_max, num=num_ribbons + 1)[1:]
    perc1 = np.percentile(y, lower, axis=0)
    perc2 = np.percentile(y, upper, axis=0)

    # fix some kwargs:
    alpha = kwargs.pop("alpha", 1)
    alpha = alpha / num_ribbons
    kwargs["alpha"] = alpha
    color = kwargs.pop("color", None)
    if color is None:
        color = ax._get_patches_for_fill.get_next_color()
    color = mcolors.to_rgba_array(color)
    kwargs["color"] = color
    container_label = kwargs.pop("label", None)

    
    polygon_collections = []
    for p1, p2 in zip(perc1, perc2):
        collection = ax.fill_between(x, p1, p2, label="_nolegend_", **kwargs)
        polygon_collections.append(collection)

    container = RibbonsContainer(polygon_collections, label=container_label)
    ax.add_container(container)
    return ax
```

```python
ax = plot_ribbons(timesteps, line_samples, label='hello')
col = ax.containers[0].collections[0]
plt.legend()
```
