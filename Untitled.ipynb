```python
import numpy as np
import matplotlib.pyplot as plt
from bizplots import plot_quantiles, plot_spaghetti, handler_map
```

```python
labels = np.random.choice(5, size=1000)
values = np.random.normal(size=1000)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5), dpi=300)
plot_quantiles(labels, values, pct_inner=25, alpha=1, ax=axes[1], label='goodbye')
plot_quantiles(labels - 1, values, alpha=1, pct_outer=75, ax=axes[0], color='C1', orientation='vertical', label='hello')
axes[0].axhline(0, linestyle=":", label='zero')
axes[0].legend(handler_map=handler_map)
axes[1].legend(handler_map=handler_map)
plt.show()
```

```python
timesteps = np.linspace(0, 10, num=1000)
a_samples = np.random.normal(0, 0.2, size=400)
b_samples = np.random.normal(loc=1, scale=0.1, size=400)

line_samples = a_samples[:, None] + b_samples[:, None] * timesteps
```

```python
plot_spaghetti(timesteps, line_samples, color='k', alpha=0.4, label='spag')
plt.legend(handler_map=handler_map)
```
